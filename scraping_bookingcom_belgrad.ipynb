```python
url_list = ['https://www.booking.com/searchresults.tr.html?ss=Belgrad&ssne=Belgrad&ssne_untouched=Belgrad&label=gen173rf-1BCAEoggI46AdIM1gDaOQBiAEBmAEouAEXyAEM2AEB6AEBiAIBogINcHJvamVjdHByby5pb6gCA7gClNjBoAbAAgHSAiQ2OTVmZWQ2My1jMDcxLTQwOTEtYjZjMC0zZjZhMjIwNDM5YjDYAgXgAgE&sid=00908ca4bb723e389dea8cabf3a08eaf&aid=304142&lang=tr&sb=1&src_elem=sb&src=searchresults&dest_id=-74897&dest_type=city&checkin=2023-04-12&checkout=2023-04-13&group_adults=2&no_rooms=1&group_children=0&sb_travel_purpose=leisure']
for i in range(1, 34):
    add = '&offset=' + str(i*30)
    url  = 'https://www.booking.com/searchresults.tr.html?ss=Belgrad&ssne=Belgrad&ssne_untouched=Belgrad&label=gen173rf-1BCAEoggI46AdIM1gDaOQBiAEBmAEouAEXyAEM2AEB6AEBiAIBogINcHJvamVjdHByby5pb6gCA7gClNjBoAbAAgHSAiQ2OTVmZWQ2My1jMDcxLTQwOTEtYjZjMC0zZjZhMjIwNDM5YjDYAgXgAgE&sid=00908ca4bb723e389dea8cabf3a08eaf&aid=304142&lang=tr&sb=1&src_elem=sb&src=searchresults&dest_id=-74897&dest_type=city&checkin=2023-04-12&checkout=2023-04-13&group_adults=2&no_rooms=1&group_children=0&sb_travel_purpose=leisure' + add
    url_list.append(url)
```

```python

title_list = []
location_list = []
ratings_list = []
ratings_explanation= []
review_counts = []
price_list = []
additional_fee = []
link_list = []

for i in range(0, len(url_list)):
    ua = UserAgent()
    header = {'User-Agent':str(ua.chrome)}
    htmlContent = requests.get(url_list[i], headers=header)

    soup= BeautifulSoup(htmlContent.content, 'html.parser')
    for item in soup.find_all('div', {'data-testid': 'property-card'}):
        try:  
            title = item.find('div', class_='fcab3ed991').get_text(strip=True)
            title_list.append(title)
        except AttributeError:
            title_list.append('not available')
        try:  
            location = item.find('span', {'data-testid': 'address'}).get_text(strip=True)
            location_list.append(location)
        except AttributeError:
            location_list.append('not available')
        try:
            rating = item.find('div', class_='d10a6220b4').get_text(strip=True).replace(',', '.')
            ratings_list.append(float(rating))
        except AttributeError:
            ratings_list.append('not rated yet')
        try:  
            rate = item.find('div', class_= 'e46e88563a').get_text(strip=True)
            ratings_explanation.append(rate)
        except AttributeError:
            ratings_explanation.append('no rating yet')
        try:    
            review = item.find('div', class_='db63693c62').get_text(strip=True).split(' ')[0]
            review_counts.append(review)
        except AttributeError:
            review_counts.append('0')
        try:  
            price = item.find('span', {'data-testid': 'price-and-discounted-price'}).get_text(strip=True)
            price_list.append((price))
        except AttributeError:
            price_list.append('not available')

        try:  
            fee = item.find('div', {'data-testid': 'taxes-and-charges'}).get_text(strip=True)
            additional_fee.append((fee))
        except AttributeError:
            additional_fee.append('not available')
        try:  
            link = item.find('a')['href']
            link_list.append((link))
        except AttributeError:
            link_list.append('not available')

        
```

```python
prices = []
for item in price_list:
    item = item[3:]
    prices.append(item)

prices
```

```python
dictionary = {'Title': title_list, 'Location': location_list, 'Rating': ratings_list, 'Rating Category': ratings_explanation, 'Number of Reviews': review_counts, 'Price for Two Adults': prices, 'Tax and other charges': additional_fee, 'Link': link_list}
df = pd.DataFrame.from_dict(dictionary)
```
